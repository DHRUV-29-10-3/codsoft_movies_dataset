```python
import pandas as pd
```

```python
df = pd.read_csv("IMDb Movies India.csv",encoding='latin1')
```

```python
df.head()
```

```python
missing_values = df.isnull().sum()
```

```python
null_values = pd.DataFrame(missing_values, columns=['missing_count'])
```

```python
null_values['missing_%'] = null_values.apply(lambda x : x/df.shape[0]*100)
```

```python
null_values
```

```python
df.info()
```

### there are two many missinf values in dataset we must remove them

```python
df = df.dropna().reset_index(drop = True)
```

```python
df['Duration'] = df['Duration'].str.replace('min','').str.strip()
df['Duration'] = pd.to_numeric(df['Duration'])
```

```python
df['Votes'] = df['Votes'].str.replace(',','').str.strip()
df['Votes'] = pd.to_numeric(df['Votes'])
```

```python
df.drop_duplicates(subset=["Name","Year"],keep = 'first', inplace=True)
```

```python
df
```

```python
df['Year'] = df['Year'].str.strip('()').astype(int)
```

```python
df['Main_Genre'] = df['Genre'].str.split(',').str[0]
```

```python
df.columns
```

### Categorical plot

```python
df.columns
```

```python
plt.figure(figsize = (7,7))
ax = sns.countplot(data = df, x = 'Main_Genre')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

```python
df['Main_Genre'].value_counts()
```

```python
df['Director'].value_counts().head(10)
```

```python
values_top_director = df['Director'].value_counts().head(10).to_frame().reset_index()['Director'].to_list()
```

```python
top_director = df[df['Director'].isin(values_top_director)]
```

```python
plt.figure(figsize = (7,7))
ax = sns.countplot(data = top_director, x = 'Director')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

```python
df['Actor 1'].value_counts().head(10)
```

```python
values_top_actor = df['Actor 1'].value_counts().head(10).to_frame().reset_index()['Actor 1'].to_list()
```

```python
top_actor = df[df['Actor 1'].isin(values_top_actor)]
```

```python
plt.figure(figsize = (7,7))
ax = sns.countplot(data = top_actor, x = 'Actor 1')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```
